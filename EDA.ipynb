2. df['diet'] - simplified to 6 categories ('Anything','Vegetarian','Other','Vegan','Halal','Kosher')
3. df['body_type'], df['job']  NaN values filled with 'rather not say'
df['drinks'], df['drugs'] NaN values filled with 'unknown'
4. df['education'] was transformed into 2 columns
first one show the level that person have is/was working on and second df['education_status'] show if person is actually studying or was dropped of or graduated;
5.  df['ethnicity'] Simplified, cause we had too many options.
So a) values  'black, other', 'asian, other', 'white,other' and others were changed as black,asian and white. 
b) values with a lot of different options were grouped as mixed
6. df['height'] -  metrics were changed to cm and outliers <135 cm and >210;
7. df['last_online'] no changes
8. df['pets']  was transformed as binary features of preferences (for ML)
df['likes_dogs'] = df['pets'].apply(lambda x: 1 if 'likes dogs' in x else 0)
df['likes_cats'] = df['pets'].apply(lambda x: 1 if 'likes cats' in x else 0)
df['has_dogs'] = df['pets'].apply(lambda x: 1 if 'has dogs' in x else 0)
df['has_cats'] = df['pets'].apply(lambda x: 1 if 'has cats' in x else 0)
df['dislikes_dogs'] = df['pets'].apply(lambda x: 1 if 'dislikes dogs' in x else 0)
df['dislikes_cats'] = df['pets'].apply(lambda x: 1 if 'dislikes cats' in x else 0)
9. df['offspring'] was changed into 2 columns  df['Kids'] and df['Want kids']
10. df['speaks'] was transofred into boolean-columns that show most used languages  [English_Proficiency,Spanish_Proficiency, French_Proficiency, German_Proficiency, Chinese_Proficiency,Italian_Proficiency,Russian_Proficiency, Hindi_Proficiency, Portuguese_Proficiency, Hebrew_Proficiency]
11. df['income'] transformed to df['income_category'] where 0 means - dont want to share data (biggest group) and groips from 1 to 12 are categories from lowest to biggest.
12. df[religion] was simplified and created also  df['religiosity_level'] to show a personal attitude to religion ('not_serious'laughing', 'somewhat_serious', 'very_serious')
13. df['sign'] was transformed into df['astrology'] that same as df['religiosity_level'] show  the attitude to astrology ('not_serious','laughing',fun to think about" ,'serious')

```python
import pandas as pd
import numpy as np
```

```python
original = pd.read_csv('okcupid_profiles.csv')
df = pd.read_csv('okcupid_profiles_cleaned.csv')
```

```python
df
```

```python
df.info()
```

```python
df['height_cm'] = df['height'] * 2.54
df.drop(df[(df['height_cm'] < 135) | (df['height_cm'] > 210)].index, inplace=True)
```

```python
df.drop(df[df['age'] > 80].index, inplace=True)
```

```python
df['body_type'].fillna('rather not say', inplace=True)
df['drinks'].fillna('unknown', inplace=True)
df['drugs'].fillna('unknown', inplace=True)
```

```python
df['diet'] = df['diet'].apply(lambda x: 'Other' if pd.isna(x) else
                              'Anything' if x in ['mostly anything', 'anything', 'strictly anything'] else
                              'Vegetarian' if x in ['mostly vegetarian', 'strictly vegetarian', 'vegetarian'] else
                              'Other' if x in ['strictly other', 'other', 'mostly other'] else
                              'Vegan' if x in ['vegan', 'strictly vegan', 'mostly vegan'] else
                              'Halal' if x in ['strictly halal', 'halal', 'mostly halal'] else
                              'Kosher' if x in ['strictly kosher', 'kosher', 'mostly kosher'] else x)
```

```python
df.info()
```

```python
df['sign'].unique()
```

```python
df['astrology'] = df['sign'].apply(lambda x: 'not_serious' if pd.notna(x) and "doesn" in x else
                                               'laughing' if pd.notna(x) and "fun to think about" in x else
                                               'serious' if pd.notna(x) and 'it matters a lot' in x else 'unknown')
```

```python
df['religion_new'] = df['religion'].apply(lambda x: x.split()[0] if isinstance(x, str) and x.split()[0] != 'other' else 'other_religion' if isinstance(x, str) else np.nan)
```

```python
df['religiosity_level'] = df['religion'].apply(lambda x: 'not_serious' if pd.notna(x) and 'not too serious' in x else
                                               'laughing' if pd.notna(x) and 'laughing about it' in x else
                                               'somewhat_serious' if pd.notna(x) and 'somewhat serious about it' in x else
                                               'very_serious' if pd.notna(x) and 'very serious about it' in x else 'unknown')
```

```python
df['education_status'] = df['education'].apply(lambda x: 'studying' if pd.notna(x) and 'working on' in str(x) else
                                                'dropped out' if pd.notna(x) and 'dropped out' in str(x) else
                                                'Graduated' if pd.notna(x) and 'graduated' in str(x) else 'unknown')
df['education'] = df['education'].str.split('from|of|on').str[-1].str.strip()
```

```python
df['likes_dogs'] = df['pets'].apply(lambda x: 1 if pd.notna(x) and 'likes dogs' in x else 0)
df['likes_cats'] = df['pets'].apply(lambda x: 1  if pd.notna(x) and 'likes cats' in x else 0)
df['has_dogs'] = df['pets'].apply(lambda x: 1 if pd.notna(x) and 'has dogs' in x else 0)
df['has_cats'] = df['pets'].apply(lambda x: 1 if pd.notna(x) and 'has cats' in x else 0)
df['dislikes_dogs'] = df['pets'].apply(lambda x: 1 if pd.notna(x) and 'dislikes dogs' in x else 0)
df['dislikes_cats'] = df['pets'].apply(lambda x: 1 if pd.notna(x) and 'dislikes cats' in x else 0)
```

```python
nan_mask = df['offspring'].isna()

df['Kids'] = df['offspring'].apply(lambda x: 0 if pd.isna(x) else 
                                   0 if 'doesn\'t have kids' in x or 'has no kids' in x or 'doesn\'t want kids' in x else 
                                   1 if 'has a kid' in x or 'have one' in x else 
                                   2 if 'has kids' in x and 'doesn\'t want more' not in x else None)

df['Want kids'] = df['offspring'].apply(lambda x: None if pd.isna(x) else 
                                        'Yes' if 'wants them' in x or 'might want more' in x or 'might want kids' in x else 
                                        'No' if 'doesn\'t want any' in x or 'doesn\'t want more' in x or 'doesn\'t want kids' in x else 
                                        'Maybe' if 'but wants them' in x or 'but might want them' in x else None)
```

```python
speaks_data = df['speaks'].dropna()  # Drop NaN values

all_languages = ', '.join(speaks_data) # Concatenate all strings in the 'speaks' column
# Split the concatenated string into a list of languages
languages_list = [language.strip() for language in all_languages.split(',')]

# Create a DataFrame to count occurrences of each language
languages_df = pd.DataFrame(languages_list, columns=['Language'])
top_20_languages = languages_df['Language'].value_counts().head(50).reset_index()

top_20_languages.columns = ['Language', 'Count']
top_20_languages
```

```python
languages_list = ['english', 'spanish', 'french', 'german', 'chinese', 'italian', 'russian', 'hindi', 'portuguese', 'hebrew']

# Iterate over languages and create a column
for language in languages_list:
    column_name = f'{language.capitalize()}_Proficiency'
    df[column_name] = df['speaks'].apply(lambda x: 'Yes' if language in str(x).lower() else 'No')
display(df)
```

```python
df['job'].fillna('rather not say', inplace=True)
df['job'].value_counts()
```

```python
ethnicity_counts = df['ethnicity'].value_counts()
display(ethnicity_counts.head(50))
```

```python
df['ethnicity'] = df['ethnicity'].replace('white, other', 'white') # Replace 'white, other' with 'white'
df['ethnicity'] = df['ethnicity'].replace('black, other', 'black') 
df['ethnicity'] = df['ethnicity'].replace('asian, other', 'asian')
```

```python
df['ethnicity'].fillna('unknown', inplace=True)
df['ethnicity'] = df['ethnicity'].apply(lambda x: 'mixed' if x.count(',') >= 2 else x)
```

```python
df['ethnicity'] = df['ethnicity'].str.replace(', other', '')
```

```python
df['ethnicity'] = df['ethnicity'].apply(lambda x: 'mixed' if x.count(',') >= 1 else x) #important to do it twice and in this order
```

```python
df['ethnicity'].value_counts()
```

```python
df['smokes'].fillna('unknown', inplace=True)
df['smokes'].value_counts()
```

```python
df['income'] = df['income'].replace(-1, 0)
df['income'] = pd.Categorical(df['income'], ordered=True)
df['income_category'] = df['income'].cat.codes
df['income_category'].value_counts()
```

```python
df = df.drop(['height', 'religion', 'sign', 'offspring', 'pets'], axis=1)
```

```python
df['religion_new'].fillna('unknown', inplace=True)
df.rename(columns={'religion_new': 'religion'}, inplace=True)
```

```python
df.dropna(subset=['height_cm'], inplace=True)
```

```python
df = df.drop(['income', 'speaks', 'location'], axis=1)
```
